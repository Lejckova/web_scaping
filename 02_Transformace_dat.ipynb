```python
import pandas as pd
import re
```

```python
data_raw = pd.read_csv('../../CodersLab_Workshop/CodersLab_Workshop/data/interim/job_offer.csv', sep=',', decimal='.', encoding='UTF-8')
data_raw.head()
```

```python
# Z datové sady vyřaďte nabídky, které nemají částku uvedenou v PLN,
data = data_raw.query("Currency == 'PLN'")
data['Currency'].unique()
data.head()
```

```python
# Změňte názvy nabídek práce a měst ( name a location_citysloupců) na malá písmena,
# .lower() přepíše na malá písmena
data['Name'] = data['Name'].apply(lambda x: x.lower())
data['Location'] = data['Location'].apply(lambda x: x.lower())
data.head(3)
```

```python
# Přidat nový salary_avg sloupec jako průměr sloupců salary_high a salary_low,
data['Salary_avg'] = (data['Salary Low'] + data['Salary High']) / 2
data.head(3)
```

Sjednotit názvy měst:
např. wroclove, wroclaw, wrocławby se mělo změnit na wrocław,  
např. krakow, kraków +1, kraków +2by se mělo změnit na kraków,  
všechny ostatní případy, které se objeví, by měly být také vyčištěny .  

```python
data['Location'].unique()
```

```python
#  .isin   - zda řádky nabývají některou z vybraných hodnot.
data.loc[data['Location'].isin(['warszawa']), 'Location'] = 'warsaw'
data['Location'].unique()
```

NoFluffJobs nesdílí informace o zemi zaměstnání, takže location_city je třeba vytvořit nebo vyplnit sloupec:  
pokud je práce vzdálená, nastavte N/A,  
jinak, pokud název města nenaznačuje jinak, nastavte PL,  

```python
# odstranit remote
data['location_country'] = 'PL'

data.loc[data['Location'] == 'remote', 'location_country'] = 'N/A'

data['location_country'].unique()



```

Přidejte nový sloupec is_senior, který informuje, zda se jedná o pozici senior či nikoliv. K tomu použijte název pozice: např Senior Data Analyst -> is_senior = 1.  
Chcete-li zkontrolovat název pozice pro slovo „Senior“, použijte regulární výrazy a re.matchmetodu.

```python
data['is_senior'] = 0
data.loc[data['Name'].apply(lambda x: True if re.search('senior', x) else False),'is_senior'] = 1

data['is_senior'].unique()
```

Uložte výsledky do data\processed\adresáře pomocí ;oddělovače, UTF-8kódování a bez indexu ( index=False). Použijte stejný název souboru jako v úloze 2, tj job_offers_yyyy_mm_dd.csv.

pozice Analytik, Engineer a Scientist

```python
data['pozice'] = 0
data.loc[data['Name'].apply(lambda x: True if re.search('scientist', x, re.IGNORECASE) else False), 'pozice'] = 'scientist'
data.loc[data['Name'].apply(lambda x: True if re.search('analyst', x, re.IGNORECASE) else False), 'pozice'] = 'analyst'
data.loc[data['Name'].apply(lambda x: True if re.search('engineer', x, re.IGNORECASE) else False), 'pozice'] = 'engineer'


data['pozice'].unique()
```
